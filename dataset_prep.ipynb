```python
import shutil

source_dir = 'Dataset'
honks_dir = os.path.join(source_dir, 'Honks')
non_honks_dir = os.path.join(source_dir, 'Non-Honks')

# Create the target directories if they don't exist
os.makedirs(honks_dir, exist_ok=True)
os.makedirs(non_honks_dir, exist_ok=True)

files = os.listdir(source_dir)
```

```python
for file in files:
    # Skip directories
    if os.path.isdir(os.path.join(source_dir, file)):
        continue
    
    # Determine the target directory based on the file name
    if file.endswith(('S.wav', 'S_.wav', 'V.wav', 'V_.wav','L.wav', 'L_.wav')):
        target_dir = honks_dir
    else:
        target_dir = non_honks_dir
    
    # Move the file
    shutil.move(os.path.join(source_dir, file), os.path.join(target_dir, file))

# Verify the results
print('Files in Honks:', os.listdir(honks_dir))
print('Files in Non-Honks:', os.listdir(non_honks_dir))
```
